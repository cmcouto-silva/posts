# Coding

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load iris dataset
df = sns.load_dataset('iris')
df.head()
```

```python
# Plot without customization
plt.figure(figsize=(6,4))
plt.title('Iris sepal vs petal length')
sns.scatterplot(data=df, x='petal_length', y='sepal_length', hue='species');
```

```python
# Customize your theme, please see:
# https://matplotlib.org/stable/tutorials/introductory/customizing.html

sns.set_theme(
    context='talk',
    style='ticks',
    font_scale=.75,
    palette='viridis',
    rc={
        'figure.figsize': (6,4),
        'axes.grid': True,
        'grid.alpha': .3,
        'axes.titlesize': 'large',
        'axes.titleweight': 'bold',
        'axes.titlepad': 30,
        'axes.labelpad': 15
    }
)

# A dictionary for customizing scatter plots
scatter_kwargs = dict(palette='bright', alpha=0.8, linewidth=0)
```

```python
# Plot with customization
sns.scatterplot(data=df, x='petal_length', y='sepal_length', hue='species', **scatter_kwargs)
plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.2), ncol=3, frameon=False)
plt.title('Iris sepal vs petal length', pad=50)
sns.despine(trim=True, offset=5)
plt.show()
```
